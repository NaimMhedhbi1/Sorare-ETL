```python
import requests
import json
import pandas as pd
import multiprocessing
```

```python
SINGLE_LEAGUE_QUERY = '''
{
  competition(slug:"k-league-1-kr"){
    name

    released
    displayName
    format
    slug
    id
    pictureUrl
    
    clubs(first:5){
    
    
    
      nodes{
      name
      players{
        nodes{
      age
      displayName 
        draftValue
          allSo5Scores{nodes{playerGameStats{minsPlayed
            so5Score{score}
          
          game{date
          awayGoals
          homeGoals
          so5Fixture{gameWeek}}}}}}}}}}}
                 
'''
```

```python
# Test request on Sorare API.
url = 'https://api.sorare.com/graphql/'
r = requests.post(url, json={'query': SINGLE_LEAGUE_QUERY})
print('request status:', r.status_code)
data = r.json()
r.json()
```
