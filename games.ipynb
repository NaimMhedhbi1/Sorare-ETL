```python
import requests
import json
import pandas as pd
import multiprocessing
```

```python
SINGLE_LEAGUE_QUERY = '''
{
  competition(slug:"k-league-1-kr"){
    name   
    released
    displayName
    format
    slug
    id
    pictureUrl   
    clubs{    
      nodes{
      name
      players{
        nodes{
      age
      displayName 
        draftValue         
          so5Scores(last:50){score}
 
          activeClub{games(startDate:"2021-10-01",endDate:"2021-10-31")
      {
        nodes{   
    date
    scored
    status
    homeGoals
    awayGoals
    so5Fixture{
      gameWeek
    }}}}}}}}}}                 
'''
```

```python
# Test request on Sorare API.
url = 'https://api.sorare.com/graphql/'
r = requests.post(url, json={'query': SINGLE_LEAGUE_QUERY})
print('request status:', r.status_code)
data = r.json()
r.json()
```

```python
for element in data['data']['competition']['clubs']: 
  #print(element['name'])
  print(element)
```

```python
fields_to_add = pd.DataFrame()
player_name = list()
player_slug = list()
player_lastFiveSo5Appearances = list()
player_lastFiveSo5AverageScore =list()
player_lastFifteenSo5Appearances = list()
player_lastFifteenSo5AverageScore = list()
clubs_name = list()
current_price = []
clubs_id = list()
competition_name = [] 
competition_foramt = [] 
competition_slug = []
competition_released = [] 
competition_id = []
competition_displayName =  [] 
players = []
```

```python
for element in data['data']['competition']['clubs']['nodes']:
  print(element) 
```

```python
for element in data['data']['competition']['clubs']['nodes']:
  print(element) 
  try:


    clubs_name.append(element['name'])
    players.append(element['players']['nodes'])

    
    
  except:
    clubs_name.append('None')
    players.append('None')


```

```python
fields_to_add['clubs_name'] = clubs_name
fields_to_add['players'] = players

fields_to_add
```

```python
fields_to_add = fields_to_add.explode('players')

fields_to_add
```

```python
df3 = df2['activeClub'].apply(pd.Series)
df4 = pd.concat([df3, df2], axis=1)
df4 = df4.drop('activeClub', 1)
#df2 = df2.drop('openAuction_name', 1)

df5 = df4['games'].apply(pd.Series)
df6 = pd.concat([df4, df5], axis=1)
df6 = df6.drop('games', 1)


df7 = df6.explode('so5Scores') 
df8 = df7.explode('nodes') 

df8
```

```python
df11 = df10['so5Fixture'].apply(pd.Series)
df12 = pd.concat([df11, df10], axis=1)
df12 = df12.drop('so5Fixture', 1)
df12
```
