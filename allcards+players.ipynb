```python
import time
import json
import pandas as pd
import requests
import multiprocessing


has_next_page = True
count = 1

df = pd.DataFrame()
player_id = []
card_rarity =[]
card_id = []
count_no_common = 0

while has_next_page:

  print(count)
  if count == 1:

        
    query1 = """
        {
          allCards(first : 20) {
            
            pageInfo {
              hasNextPage
              hasPreviousPage
              startCursor
              endCursor
            }
            nodes {
              rarity
              player {
                id
              }

              }
            }
          }
        
        """
  else:


        
    query1 = '''
        {
          allCards(after: "''' + last_cursor_ + '''") {
            pageInfo {
              hasNextPage
              hasPreviousPage
              startCursor
              endCursor
            }
            nodes {
              id 
              rarity
              player {
                id
              }

              }
            }
          }
        
        '''
  url = 'https://api.sorare.com/graphql/'
  r = requests.post(url, json={'query': query1})
  time.sleep(0)
  data = r.json()

  print(data['data']['allCards']['pageInfo']['hasPreviousPage'])
  print(data['data']['allCards']['pageInfo']['hasNextPage'])
  last_cursor_ = data['data']['allCards']['pageInfo']['endCursor']
  has_next_page = data['data']['allCards']['pageInfo']['hasNextPage']
  print(last_cursor_)

  for element in data['data']['allCards']['nodes']:


    if element['rarity'] is not None:
      count_no_common = count_no_common + 1


      

            
      try : 

        player_id.append(element['player']['id'])
        card_rarity.append(element['rarity'])
        card_id.append(element['id'])
      except : 
        player_id.append('0')
        card_rarity.append('None')  
        card_id.append('None')            



  count = count + 1
  print(count)
  print('count_no_common: ', count_no_common)


```
