```python
import time
import json
import pandas as pd
import requests
import multiprocessing


has_next_page = True
count = 1

df = pd.DataFrame()
player_id = []
card_rarity =[]

count_no_common = 0

while has_next_page:
    print(count)
    if count == 1:
        query1 = """
        {

        allCards(first : 20) {
            
            pageInfo {
              hasNextPage
              hasPreviousPage
              startCursor
              endCursor
            }
        edges {
        node {
        id
        gameForLeague{date}
        xpAfterTransfer
        player{
          id
          displayName}
        price
        
        
      }
    }
  }
}  
        """
    else:
        query1 = '''
         {

        allCards(after: "''' + last_cursor_ + '''") {
            
            pageInfo {
              hasNextPage
              hasPreviousPage
              startCursor
              endCursor
            }
        edges {
        node {
        id
        gameForLeague{date}
        xpAfterTransfer
        player{
          id
          displayName}
        price
        
        
      }
    }
  }
}  
        
        '''
    url = 'https://api.sorare.com/graphql/'
    r = requests.post(url, json={'query': query1})
    time.sleep(0)
    data = r.json()

    #print(data['data']['allCards']['pageInfo']['hasPreviousPage'])
    #print(data['data']['allCards']['pageInfo']['hasNextPage'])
    last_cursor_ = data['data']['allCards']['pageInfo']['endCursor']
    has_next_page = data['data']['allCards']['pageInfo']['hasNextPage']
    print(last_cursor_)

    for element in data['data']['allCards']['node']:
      
        if element['palyer'] is not None:
            count_no_common = count_no_common + 1
            try : 
              player_id.append(element['player']['id'])
              card_rarity.append(element['player'])
            except : 
              player_id.append('0')
              card_rarity.append('None')              



    count = count + 1
    print('count_no_common: ', count_no_common)

df['player_id'] = player_id
```
