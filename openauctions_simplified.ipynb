```python
import requests
import json
import pandas as pd
import multiprocessing
```

```python
SINGLE_LEAGUE_QUERY = '''
  {
competition(slug :"k-league-1-kr")
    { 
      clubs
      {
        nodes{
          name
    openAuctions
          {
            nodes{     
      name
      endDate
      startDate
      currentPrice 
      cards {
        rarity

        price
        name
 
        player 
        {
 
          age
          position
            displayName
            id
          }
        }      
  }
  }
  } 
}
  }
}
 
'''
```

```python
# Test request on Sorare API.
url = 'https://api.sorare.com/graphql/'
r = requests.post(url, json={'query': SINGLE_LEAGUE_QUERY})
print('request status:', r.status_code)
data = r.json()
r.json()
```
