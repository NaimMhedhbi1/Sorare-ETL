```python
import requests
import json
import pandas as pd
import multiprocessing
```

```python
SINGLE_LEAGUE_QUERY = '''

{
  transferMarket {
    auctions{
      
      nodes {
        currentPrice
        
        card {
          rarity
          name
          price

          player {
            
	        status{
              playingStatus
            }
            age
            position

            so5Scores(last:100){
              score
            }
            activeClub {
              name
              domesticLeague {
                name
              }
            }
            displayName
            id
          }
        }
        endDate
        bestBid {
          amount
        }
      }
    }
  }
}
 
'''
```

```python
# Test request on Sorare API.
url = 'https://api.sorare.com/graphql/'
r = requests.post(url, json={'query': SINGLE_LEAGUE_QUERY})
print('request status:', r.status_code)
data = r.json()
r.json()
```
